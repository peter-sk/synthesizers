# Use Case 1
1. Load dataset from local directory
2. Generate and evaluate synthetic data
3. Save synthetic data as an Excel file (.xlsx)

## Imports

```python
from synthesizers import Load
```

## Load and generate data
* `split_size`:         Split the loaded data into training and holdout data. Holdout data is used for evaluation.
* `train_plugin`:       Passes the argument `plugin` to the train adapter.
* `gen_count`:          Number of synthetic records to generate.
* `eval_target_col`:    Attribute name for an attribute that is used for supervised evaluation methods.
* `save_name`:          Path and file name for the output file. The file extension determines the output type.
* `save_key`:           The generation process produces a dictionary-like output. The `save_key` is dictionary key for which data is saved.

```python
Load(r"data/Wine_Quality_Data.csv").Synthesize(split_size=0.8, train_plugin="bayesian_network", gen_count=1e4, eval_target_col="color", save_name="synthetic_wine.xlsx", save_key="synth")
```

```python
state = Load("mstz/breast").Split(size=0.8).Train(train_adapter="synthpop").Generate(count=100).Evaluate(target_col="is_cancer").Save(name="wine.xlsx", key="synth")
```
